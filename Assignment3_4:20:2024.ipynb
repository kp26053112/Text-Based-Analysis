```python
import pandas as pd
file = "C:\\Users\\Dell\\OneDrive\\Desktop\\sem1\\big data\\script\\nba2k20-full.csv"
df = pd.read_csv(file)
df.head(20)
```

```python
import nltk
nltk.download('stopwords')
stopwords = set(nltk.corpus.stopwords.words('french'))
print(stopwords)
```

```python
import regex as re
def tokenize(text):
    return re.findall(r'[\w-]*[a-z|A-Z][\w-]*', text)
def remove_stop(tokens):
    return [t for t in tokens if t.lower() not in stopwords]
pipeline = [str.lower, tokenize, remove_stop]
def prepare(text, pipeline):
    tokens = text
    for fun in pipeline:
        tokens = fun(tokens)
    return tokens
df['tokens'] = df['full_name'].apply(prepare, pipeline=pipeline)
```

```python
from collections import Counter
counter = Counter()
df['tokens'].map(counter.update)
freq_df = pd.DataFrame.from_dict(counter, 
orient='index', columns=['freq'])
freq_df = freq_df.query('freq >= 1')
freq_df.index.name = 'tokens'
freq_df = freq_df.sort_values('freq', ascending=False)
freq_df.head(10)
```

```python
from wordcloud import WordCloud
from matplotlib import pyplot as plt
wc = WordCloud(background_color='blue', width = 300, height=300, margin=2)
wc.fit_words(freq_df['freq'].to_dict() )
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wc)
plt.axis("off")
plt.tight_layout(pad = 0)
plt.show()
```
